```python
import fitz
from PIL import Image
import io
```

```python
pdf_file = fitz.open(r"C:\Users\DELL\Desktop\Vendor Bot SoW V10.pdf")
```

```python
# Iterate over all the pdf files pages
for pages_index in range (len(pdf_file)):
    # Assigning a variable "page" to each page of pdf.
    page = pdf_file[pages_index]
    # Get the image list
    image_list = page.get_images()
    # Iterating over the image list.
    for image_index , img in enumerate(image_list,start = 1):
        xref = img[0]
          
        # extract the image bytes
        base_image = pdf_file.extract_image(xref)
        image_bytes = base_image["image"]
          
        # get the image extension
        image_ext = base_image["ext"]
        # Load it to PIL
        image = Image.open(io.BytesIO(image_bytes))
        # Save
        image.save(open(f"image{pages_index+1}_{image_index}.{image_ext}", "wb"))
    
```
